```python
import pandas as pd

df = pd.read_csv("tickets.csv")
df.head()
```

```python
def clean_message(text):
    text = text.lower()     # convert to lowercase
    text = text.replace(".", "")
    text = text.replace(",", "")
    text = text.replace("!", "")
    text = text.replace("?", "")
    return text

df["clean_message"] = df["Ticket Description"].apply(clean_message)
```

```python
def find_issue(message):
    if "payment" in message or "card" in message:
        return "Payment Issue"
    elif "login" in message or "password" in message:
        return "Login Issue"
    elif "delivery" in message or "shipping" in message:
        return "Delivery Issue"
    elif "refund" in message:
        return "Refund Issue"
    elif "error" in message or "bug" in message:
        return "Technical Issue"
    else:
        return "General Issue"

df["Issue Type"] = df["clean_message"].apply(find_issue)
```

```python
def assign_priority(message):
    if "urgent" in message or "immediately" in message:
        return "P0"
    elif "asap" in message or "soon" in message:
        return "P1"
    elif "delay" in message:
        return "P2"
    else:
        return "P3"

df["Priority"] = df["clean_message"].apply(assign_priority)
```

```python
def assign_sla(priority):
    if priority == "P0":
        return 2
    elif priority == "P1":
        return 6
    elif priority == "P2":
        return 24
    else:
        return 48

df["SLA Hours"] = df["Priority"].apply(assign_sla)
```

```python
from datetime import datetime

current_time = datetime.now()

df["SLA Due Time"] = current_time + pd.to_timedelta(df["SLA Hours"], unit="h")
```
